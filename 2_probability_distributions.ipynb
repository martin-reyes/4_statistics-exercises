1. A bank found that the average number of cars waiting during the noon hour at a drive-up window follows a Poisson distribution with a mean of 2 cars. Make a chart of this distribution and answer these questions concerning the probability of cars waiting at the drive-up window.

    - What is the probability that no cars drive up in the noon hour?
        - **.135335**
    - What is the probability that 3 or more cars come through the drive through?
        - **.32332**
    - How likely is it that the drive through gets at least 1 car?
        - **.86466**


```python
distribution = stats.poisson(2)

plt.figure(figsize=(4,2))
plt.hist(distribution.rvs(10000), bins=10,color='skyblue',
         edgecolor='black', alpha=0.7)
plt.xlabel('Cars')
plt.ylabel('Frequency')
plt.title('Histogram Plot of Poisson Distribution of Cars')
```

```python
# probability x cars show
stats.poisson(2).pmf(range(8))
```

```python
# probability <= x cars show
stats.poisson(2).cdf(range(9))
```

```python
# probability > x cars show
stats.poisson(2).sf(range(8))
```

```python
# 1. What is the probability that no cars drive up in the noon hour?
stats.poisson(2).pmf(0)
```

```python
# 2. What is the probability that 3 or more cars come through the drive through?
stats.poisson(2).sf(3-1)
```

```python
1-stats.poisson(2).cdf(2)
```

```python
# 3. How likely is it that the drive through gets at least 1 car?
stats.poisson(2).sf(0)
```

2. Grades of State University graduates are normally distributed with a mean of 3.0 and a standard deviation of .3. Calculate the following:

    - What grade point average is required to be in the top 5% of the graduating class?
        - **3.49**
    - What GPA constitutes the bottom 15% of the class?
        - **2.69**
    - An eccentric alumnus left scholarship money for students in the third decile from the bottom of their class. Determine the range of the third decile. Would a student with a 2.8 grade point average qualify for this scholarship?
        - **Yes (.252)**
    - If I have a GPA of 3.5, what percentile am I in?
        - **95th**

```python
distribution = stats.norm(3, .3)

plt.figure(figsize=(4,2))
plt.hist(distribution.rvs(10000), bins=20,color='skyblue',
         edgecolor='black', alpha=0.7)
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.title('Histogram Plot of Normal Distribution of Grades')
```

```python
print('PDF: ', distribution.pdf([2.5,3,3.5]))
print('CDF: ', distribution.cdf([2.5,3,3.5]))
print('SF: ', distribution.sf([2.5,3,3.5]))
```

```python
# What grade point average is required to be in the top 5% of the graduating class?
distribution.isf(.05)
```

```python
# What GPA constitutes the bottom 15% of the class?
distribution.ppf(.15)
```

```python
# An eccentric alumnus left scholarship money for students in the third decile from the bottom of their class. 
# Determine the range of the third decile. Would a student with a 2.8 GPA qualify?
print(distribution.ppf([.2, .3]))
distribution.cdf(2.8)
```

```python
# If I have a GPA of 3.5, what percentile am I in?
distribution.cdf(3.5)
```

3. A marketing website has an average click-through rate of 2%. One day they observe 4326 visitors and 97 click-throughs. How likely is it that this many people or more click through?

```python
distribution = stats.binom(4326, .02)

plt.figure(figsize=(4,2))
plt.hist(distribution.rvs(50000), bins=25,color='skyblue',
         edgecolor='black', alpha=0.7)
plt.xlabel('Click-Throughs')
plt.ylabel('Frequency')
plt.title('Histogram Plot of Normal Distribution of Click-Throughs')
```

```python
distribution.sf(97-1)
```

4. You are working on some statistics homework consisting of 100 questions where all of the answers are a probability rounded to the hundreths place. Looking to save time, you put down random probabilities as the answer to each question.

    - What is the probability that at least one of your first 60 answers is correct?
        - **.458**

```python
distribution = stats.binom(60, .01)
plt.figure(figsize=(4,2))
plt.hist(distribution.rvs(10000), bins=20,color='skyblue',
         edgecolor='black', alpha=0.7)
plt.xlabel('Answers Correct')
plt.ylabel('Frequency')
plt.title('Histogram Plot of Binomial Distribution of Answers Correct')
```

```python
distribution.sf(0) # P(X >= 1) or P(X > 0)
```

6. You want to get lunch at La Panaderia, but notice that the line is usually very long at lunchtime. After several weeks of careful observation, you notice that the average number of people in line when your lunch break starts is normally distributed with a mean of 15 and standard deviation of 3. If it takes 2 minutes for each person to order, and 10 minutes from ordering to getting your food, what is the likelihood that you have at least 15 minutes left to eat your food before you have to go back to class? Assume you have one hour for lunch, and ignore travel time to and from La Panaderia.

- I will have at least 15 minutes to eat if I get my food in **45** minutes or less. I will wait for 45 or less minutes if the line takes 35 minutes (10 minute cook time). The line will take 35 minutes or less if **17 or less** people are in the line.

**Find P(X<17)**

**Answer: .7475**

```python
mean_wait = 15
std_line_length = 3
# order_time_per_person = 2
# cook_time = 10
# target_time_to_get_food = 45 # or less
distribution_line_length = stats.norm(mean_wait, std_line_length)

plt.figure(figsize=(4,2))
plt.hist(distribution_line_length.rvs(10000), bins=20,color='skyblue',
         edgecolor='black', alpha=0.7)
plt.xlabel('People')
plt.ylabel('Frequency')
plt.title('Histogram Plot of Normal Distribution of People')
```

```python
distribution_line_length.cdf(16)
```

    - What percent of employees earn less than 60,000?
    - What percent of employees earn more than 95,000?
    - What percent of employees earn between 65,000 and 80,000?
    - What do the top 5% of employees make?

```python
import pandas as pd
import env
```

```python
# code takes awhile to run. Uncomment to run
sql = '''SELECT * FROM salaries WHERE to_date > NOW()'''
con = f'''mysql+pymysql://{env.user}:{env.pwd}@{env.host}/employees'''
df = pd.read_sql(sql, con)
```

```python
df.head()
```

Find the average salary of current employees, along with the standard deviation

```python
curr_salaries = df[df['to_date'].astype(str).str.startswith('9999')]
```

```python
curr_salary_mean = curr_salaries['salary'].mean()
curr_salary_std = curr_salaries['salary'].std()
curr_salaries['salary'].agg(['mean', 'std'])
```

```python
salary_dist = stats.norm(curr_salary_mean, curr_salary_std)
```

What percent of employees earn less than 60,000?

```python
salary_dist.cdf(60_000)
```

What percent of employees earn more than 95,000?

```python
salary_dist.sf(95_000)
```

What percent of employees earn between 65,000 and 80,000?

```python
salary_dist.cdf(80_000) - salary_dist.cdf(65_000)
```

What do the top 5% of employees make?

```python
salary_dist.isf(.05)
```
